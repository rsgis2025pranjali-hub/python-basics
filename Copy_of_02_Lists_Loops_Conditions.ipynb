## 6. Mini Task â€“ Rainfall classification
### ðŸ”¶ Task 6.1 â€“ Classify rainfall intensity

```python
rain = [0.0, 12.5, 34.0, 2.0, 0.0, 55.1, 8.4]

no_rain = 0.0
light = 2.0
moderate = 12.5
heavy = 55.1

for r in rain:
    if r ==0.0:
      no_rain +=1
    if r > 0 and r <= 10:
      light +=1
    if r > 10 and r <= 30:
      moderate +=1
    if r > 30:
      heavy +=1

print(f"No rain days:" , no_rain)
print(f"Light rain days:" , light)
print(f"Moderate rain days:" , moderate)
print(f"Heavy rain days:", heavy)
```
